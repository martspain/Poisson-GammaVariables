# Ejercicio 1

### Considere que usted analizará hasta un máximo de 16 huracanes. Grafique la función PMF.

```python
# Function using the Poisson distribution
def Poisson(xval, lambd):
  return (exp(-lambd) * pow(lambd, xval))/(factorial(xval))
```

```python
xmin = 0 # min 0 hurricanes
xmax = 16 # max 16 hurricanes
xlam = 7 # average hurricanes

xvalues = []
yvalues = []

for current in range(xmax + 1):
  xvalues.append(current)
  yvalues.append(Poisson(current, xlam))
```

```python
plt.plot(xvalues, yvalues, 'y--')
plt.xticks(xvalues)
plt.title("Probability Mass Function")
plt.xlabel('Number of Hurricanes')
plt.ylabel('Probability')
# plt.grid()
plt.show()
```

## Considere que usted analizará un máximo de 16 huracanes. Gráfique la función CDF.

```python
# Based on implementation made by Simon, David on this url: https://stackoverflow.com/questions/40952773/poisson-cumulative-probability-function-in-python-returns-a-close-but-not-true-v

def PoissonCDF(xmax, lambd):
  result = []
  cumulative = 0.0
  i = 0
  while i <= xmax:
    cumulative += Poisson(i, lambd)
    result.append(cumulative)
    i += 1

  return result
```

```python
yvalues = PoissonCDF(16, 7)

yvalues
```

```python
plt.plot(xvalues, yvalues, 'b--')
plt.xticks(xvalues)
plt.title("Cumulative Distribution Function")
plt.xlabel('Number of Hurricanes')
plt.ylabel('Probability')
# plt.grid()
plt.show()
```

```python
# Inverse function of Poisson distribution
def PoissonMediumTime(xval, lamval):
  if xval < 1:
    return - (log(1 - xval) / lamval)
  else: return 0
```

```python
lambdaVal = 5
maxVal = 10

xvalues = []
yvalues = []

#Set the seed so that this gives the same results always.
random.seed(12345)

for xval in range(maxVal + 1):
  xvalues.append(xval)
  yvalues.append(PoissonMediumTime(random.random(), lambdaVal))

yvalues
```

```python
print("# Pati\t|\tProb")
for i in range(len(yvalues)):
  print(str(xvalues[i]) + "\t|\t" + str(yvalues[i]))
```

```python
plt.plot(xvalues, yvalues, 'g--')
#plt.xticks(xvalues)
plt.title("Intermediate time in hours for every patient")
plt.xlabel('Patients')
plt.ylabel('Time in hours')
#plt.grid()
plt.show()
```

```python
# Gamma PDF
# x = iter, lam = rate of events, and k = number of events
def gammaPDF(x, lam, k):
  return (lam * exp(-lam * x) * pow(lam * x, k - 1)) / (factorial(k - 1))
```

```python
# Code based on implementation made by Aerin Kim, 2019. https://towardsdatascience.com/gamma-distribution-intuition-derivation-and-examples-55f407423840
def plot_gamma_lambda():
  kVal = 3  # k = 10
  x = np.linspace(0, 50, 1000)

  lambValue = 2
  y1 = gamma.pdf(x, kVal, scale=1/lambValue)

  lambValue = 1
  y2 = gamma.pdf(x, kVal, scale=1/lambValue)

  lambValue = 0.5
  y3 = gamma.pdf(x, kVal, scale=1/lambValue)

  # Plot graph with three values
  plt.title("PDF of Gamma Distribution (k = 3)")
  plt.xlabel("Time in hours")
  plt.ylabel("Probability Density")
  plt.plot(x, y1, label="λ = 2", color='gold')
  plt.plot(x, y2, label="λ = 1", color='blue')
  plt.plot(x, y3, label="λ = 0.5", color='red')
  plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=1, fontsize=12)
  plt.ylim([0, 0.60])
  plt.xlim([0, 20])
  plt.show()
```

```python
# Gamma PDF for lambda = 2

plot_gamma_lambda()
```
